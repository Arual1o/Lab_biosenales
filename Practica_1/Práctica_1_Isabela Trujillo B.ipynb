**Reto:** 
Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
def Reto(n):
    notas = n.values()
    notas2 = pd.Series(n)
    numeros = pd.to_numeric(notas2, errors='coerce')
    maximo = max(notas)
    minimo = min(notas)
    promedio = numeros.mean()
    redondeo = round(promedio,2)
    desviacion = numeros.std()
    redondeo_des = round(desviacion,3)
    serie = Series([maximo, minimo, redondeo, redondeo_des], index=["Maximo","Minimo","Media","Desviaci√≥n T√≠pica"])
    print(serie)

diccionario = {'Estudiante 1':'4.5','Estudiante 2': '4.0', "Estudiante 3": '4.8'}
Reto(diccionario)
    
```
