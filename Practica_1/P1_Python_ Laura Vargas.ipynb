Construyendo la función y[n]=sin(π*0.18n) en el intervalo 0<=n<=80

La senoidal y2[n]=cos(2π*0.03n) y la sumatoria y el producto de las dos señales anteriores

```python
#señal 1
n=np.linspace(0,80,81)
y=np.sin(np.pi*0.18*n)

#señal 2
y2=np.cos(2*np.pi*0.03*n)

#señal 3 (sumatoria de señales 1 y 2)
s=y+y2

#Señal 4 (producto entre y*y2)
t=y*y2
```

```python
#Graficación Señales y - y2
fig,ax=plt.subplots(figsize=(5,2.7), layout='constrained')
ax.plot(n,y,label='Señal 1')
ax.plot(n,y2,label='Señal 2')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.legend()
plt.show()

#Graficación señales s y t
fig,ax=plt.subplots(figsize=(5,2.7), layout='constrained')
ax.plot(n,s,color='red',label='Sumatoria')
ax.plot(n,t,color='purple',label='Producto')
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.legend()
plt.show()
```
