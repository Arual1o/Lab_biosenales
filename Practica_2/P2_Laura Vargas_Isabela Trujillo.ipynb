```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from functools import reduce
```

#### Impulso Unitario

```python
def impseq(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n-n0)==0
    return[x,n]
n0=2
n1=-2
n2=3
x,n=impseq(n0,n1,n2)
print(x,'\n',n)
plt.figure()
plt.stem(n,x)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Impulso unitario')
```

##### Escalón Unitario
n es el rango

```python
def escu(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n-n0)>=0
    x=x*1
    return[x,n] 

x,n=escu(0,-2,3)
print(x,'\n',n)
plt.figure()
plt.stem(n,x)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Escalón unitario')
```

##### Rampa

```python
def rampa(n0,n1,n2):
    n=np.arange(n1,n2+1)
    u,x=escu(n0,n1,n2)
    r=u*(n-n0)
    return[r,n]

r,n=rampa(0,-2,5)
print(r,'\n n es es: \n',n)
plt.figure()
plt.stem(n,r)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Rampita')
```

#### Punto 3.a

```python

def x1(n):
    x1, _ = impseq(-1, n, n)
    x2, _ = impseq(-3, n, n)
    x3, _ = impseq(-2, n, n)
    x4, _ = impseq(0, n, n)
    return 3*x1 + 5*x2 + 3*x3 + 3*x4 + x4

n_values = np.arange(-4, 2)
x1_values = [x1(n) for n in n_values]

plt.stem(n_values, x1_values)
plt.xlabel('n')
plt.ylabel('x')
plt.title('Punto 3.a')
plt.show()
```

#### Punto 3.d

```python

def x4(n):
    x1, _ = rampa(-3, n, n)
    x2, _ = rampa(2, n, n)
    x3, _ = escu(3, n, n)   
    return 2*x1 - 1*x2 - 5*x3
    

n_values = np.arange(-10, 10)
x4_values = [x4(n) for n in n_values]

plt.stem(n_values, x4_values)
plt.xlabel('n')
plt.ylabel('x')
plt.title('Punto 3.d')
plt.show()

```
