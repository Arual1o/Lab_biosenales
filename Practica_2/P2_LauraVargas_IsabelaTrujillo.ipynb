Operar con señales discretas. Impulso, escalón, rampa
#### *Impulso* / No tiene en cuenta la amplitud, se multiplica por x


```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt

def impseq(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n-n0)==0
    return[x,n]
n0=2
n1=-2
n2=3
x,n=impseq(n0,n1,n2)
print(x,'\n',n)
plt.figure()
plt.stem(n,x)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Impulso unitario')
```

### *Escalón*

```python
def escu(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n-n0)>=0
    x=x*1
    return[x,n] 

x,n=escu(0,-2,3)
print(x,'\n',n)
plt.figure()
plt.stem(n,x)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Escalón unitario')
```

### *Rampa*

```python
def rampa(n0,n1,n2):
    n=np.arange(n1,n2+1)
    u,x=escu(n0,n1,n2)
    r=u*(n-n0)
    return[r,n]

r,n=rampa(0,-2,5)
print(r,'\n n es es: \n',n)
plt.figure()
plt.stem(n,r)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Rampita')

# n=np.arange(n1,n2+1)
# u,n=escu(n0,n1,n2)
# x= u*(n-n0)
```
