Operar con señales discretas. Impulso, escalón, rampa
#### *Impulso* / No tiene en cuenta la amplitud, se multiplica por x


```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt

def impseq(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n-n0)==0
    return[x,n]
n0=2
n1=-2
n2=3
x,n=impseq(n0,n1,n2)
print(x,'\n',n)
plt.figure()
plt.stem(n,x)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Impulso unitario')
```

### *Escalón*

```python
def escu(n0,n1,n2):
    n=np.arange(n1,n2+1)
    x=(n-n0)>=0
    x=x*1
    return[x,n] 

x,n=escu(0,-2,3)
print(x,'\n',n)
plt.figure()
plt.stem(n,x)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Escalón unitario')
```

### *Rampa*

```python
def rampa(n0,n1,n2):
    n=np.arange(n1,n2+1)
    u,x=escu(n0,n1,n2)
    r=u*(n-n0)
    return[r,n]

r,n=rampa(0,-2,5)
print(r,'\n n es es: \n',n)
plt.figure()
plt.stem(n,r)
plt.grid()
plt.xlabel('n')
plt.ylabel('x')
plt.title('Rampita')

# n=np.arange(n1,n2+1)
# u,n=escu(n0,n1,n2)
# x= u*(n-n0)
```

### *Exponencial*

```python
def expc(n0,n1,n2):
    n=np.arange(n1,n2+1)
    u,x=escu(n0,n1,n2)
    r=u*(n-n0)
    return[r,n]

x = np.linspace(0, 10, 50)  # Valores de x
y_real = np.cos(x)  # Parte real: cos(x)
y_imag = np.sin(x)  # Parte imaginaria: sin(x)

plt.stem(x, y_real, label='Parte real',linefmt='purple')  # Grafica la parte real
plt.stem(x, y_imag, label='Parte imaginaria',linefmt='lightgreen')  # Grafica la parte imaginaria

plt.xlabel('x')  # Etiqueta del eje OX
plt.ylabel('y')  # Etiqueta del eje OY
plt.title('Gráfica de la función exponencial compleja')  # Título del gráfico
plt.legend()  # Leyenda del gráfico
plt.show()  # Muestra el gráfico
```
